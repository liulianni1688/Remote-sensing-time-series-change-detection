# 函数1：HOG 特征提取与相似矩阵显示

```python
################################## 结构信息处理
import cv2
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def hogRGB(images_R,images_G,images_B,reSize,pixels_per_cell,cells_per_block,orientations):
    _,_,imageNum = images_R.shape
    imgR = images_R[:,:,0]
    imgG = images_G[:,:,0]
    imgB = images_B[:,:,0]
    img = np.dstack((imgR,imgG,imgB))
    img = cv2.resize(img, reSize) 
    a1,bb = hog(img, orientations=orientations, pixels_per_cell=pixels_per_cell, cells_per_block=cells_per_block, block_norm='L2',multichannel=True, visualize=True,feature_vector=True)
    cc = a1 #[~np.isnan(a1)]       
    for i in range(1,imageNum):
        imgR = images_R[:,:,i]
        imgG = images_G[:,:,i]
        imgB = images_B[:,:,i]
        img = np.dstack((imgR,imgG,imgB))
        img = cv2.resize(img, reSize) 
        a1,b1 = hog(img, orientations=orientations, pixels_per_cell=pixels_per_cell, cells_per_block=cells_per_block, block_norm='L2',multichannel=True, visualize=True,feature_vector=True)
        cc = np.dstack((cc, a1)) #[~np.isnan(a1)]
        bb = np.dstack((bb, b1))
    return  cc


def similarityMatrix(compare1): #compare1为3维np数组
        row,col,imageNum = compare1.shape
        matrix = []
        for j in range(imageNum):
            base = compare1[:,:,j]
            base = base.reshape((1,row*col))
            non_nanpos1 = ~(np.isnan(base))    
            nochange_corrcoef = []
            for i in range(0,imageNum):
                base1 = base
                compare = compare1[:,:,i]
                compare = compare.reshape((1,row*col))
                non_nanpos2 = ~(np.isnan(compare))
                non_nanpos = (non_nanpos2==True) & (non_nanpos1==True)
                base1 = base1[non_nanpos]
                compare = compare[non_nanpos]
                value = np.corrcoef(np.vstack((base1,compare)))[1,0]
                nochange_corrcoef.append(value)
            matrix.append(nochange_corrcoef)
            ss = 1-np.array(matrix)
            ss[np.isnan(ss)] = 2    
            ss[np.diag_indices(ss.shape[0])] = 0
        
        return ss
```

# 函数2：图分割和合并

```python
# 最小分割
from itertools import combinations
import time
import networkx as nx

class Node:
    def __init__(self, parent, rank=0, size=1):
        self.parent = parent
        self.rank = rank
        self.size = size

    def __repr__(self):
        return '(parent=%s, rank=%s, size=%s)' % (self.parent, self.rank, self.size)

class Forest:
    def __init__(self, num_nodes):
        self.nodes = [Node(i) for i in range(num_nodes)]
        self.num_sets = num_nodes

    def size_of(self, i):
        return self.nodes[i].size

    def find(self, n):
        temp = n
        while temp != self.nodes[temp].parent:
            temp = self.nodes[temp].parent

        self.nodes[n].parent = temp
        return temp

    def findNodesbyParent(self, n):
        nodesP = []
        for i in range(len(self.nodes)):
            if self.nodes[i].parent == n:
                nodesP.append(i)
        return nodesP

    def findEdgesbySameNodes(self,edges,nodelist):
        #另前面节点编号大于后面节点编号
        sorted_graph_array = np.array(edges)
        d = sorted_graph_array[:,0] - sorted_graph_array[:,1]
        pos = np.where(d<0)
        if len(pos) >0:
            c = sorted_graph_array[pos,0]
            sorted_graph_array[pos,0] = sorted_graph_array[pos,1]
            sorted_graph_array[pos,1] = c

        #根据节点组合寻找边缘
        pa_node = sorted(nodelist,reverse=True)  #节点降序排列
        pa_comb = list(combinations(pa_node, 2)) #节点组合
        pa_edges = []
        if np.size(pa_comb) == 0: #获取节点组合对应的edge
            pa_edges = []
        else:
            for comb in pa_comb:            
                pos = np.where((sorted_graph_array[:,0:2] == np.array(comb)).all(1))[0]
                if np.size(pos) != 0 :
                    pa_edges.append(tuple(sorted_graph_array[pos,:].flatten()))
            if len(pa_edges) == 0:
                pa_edges = []
        return pa_edges

    def findEdgesbyDiffNodes(self,edges,nodelist1,nodelist2):
        #另前面节点编号大于后面节点编号
        sorted_graph_array = np.array(edges)
        d = sorted_graph_array[:,0] - sorted_graph_array[:,1]
        pos = np.where(d<0)
        if np.size(pos) >0:
            c = sorted_graph_array[pos,0]
            sorted_graph_array[pos,0] = sorted_graph_array[pos,1]
            sorted_graph_array[pos,1] = c

        #根据节点组合寻找边缘
        pa_node = sorted(nodelist1,reverse=True)  #节点降序排列
        pa_comb = list(combinations(pa_node, 2)) #节点组合

        pb_node = sorted(nodelist2,reverse=True)  #节点降序排列
        pb_comb = list(combinations(pb_node, 2)) #节点组合

        nodess = sorted(nodelist1+nodelist2,reverse=True)  #节点降序排列
        nodess_comb = list(combinations(nodess, 2)) #
        
        nodediff = list(set(nodess_comb).difference(set(pa_comb+pb_comb)))

        pa_edges = []
        if np.size(nodediff) == 0: #获取节点组合对应的edge
            pa_edges = []
        else:
            for comb in nodediff:            
                pos = np.where((sorted_graph_array[:,0:2] == np.array(comb)).all(1))[0]
                if np.size(pos) != 0 :
                    pa_edges.append(tuple(sorted_graph_array[pos,:].flatten()))
            if len(pa_edges) == 0:
                pa_edges = []
        return pa_edges

    def merge(self, a, b):
        if self.nodes[a].rank > self.nodes[b].rank:
            self.nodes[b].parent = a
            self.nodes[a].size = self.nodes[a].size + self.nodes[b].size
        else:
            self.nodes[a].parent = b
            self.nodes[b].size = self.nodes[b].size + self.nodes[a].size

            if self.nodes[a].rank == self.nodes[b].rank:
                self.nodes[b].rank = self.nodes[b].rank + 1

        self.num_sets = self.num_sets - 1

    def print_nodes(self):
        for node in self.nodes:
            print(node)

def create_edge(img, width, x, y, x1, y1, diff):
    vertex_id = lambda x, y: y * width + x
    w = diff(img, x, y, x1, y1)
    return (vertex_id(x, y), vertex_id(x1, y1), w)


def remove_small_components(forest, graph, min_size):
    for edge in graph:
        a = forest.find(edge[0])
        b = forest.find(edge[1])

        if a != b and (forest.size_of(a) < min_size or forest.size_of(b) < min_size):
            forest.merge(a, b)

    return  forest

def get_segLabel(forest): #获取各节点对应的类别
    time = 0
    time_list = []
    label_list = []
    for node in forest.nodes:
        label = node.parent
        label_list.append(label)
        time_list.append(time)
        time = time+1
    return label_list



def preprocess(forest,label_list): #去除突变信息，
    uniqVlaue = list(map(itemgetter(0), groupby(label_list)))
    all_pos = []
    uniqPoslist = []
    i = 0
    for j in range(len(label_list)):
        if label_list[j] == uniqVlaue[i]:
            uniqPoslist.append(j)
        else:
            all_pos.append(uniqPoslist)
            uniqPoslist = [] 
            i = i+1
            uniqPoslist.append(j)
    all_pos.append(uniqPoslist) 
    
    if len(all_pos)>=2:
        for i in range(1,len(all_pos)-1):
            if (len(all_pos[i]) <=1) & ((label_list[all_pos[i-1][-1]]) == (label_list[all_pos[i+1][0]])):
                 forest.nodes[all_pos[i][0]].parent = label_list[all_pos[i-1][-1]]    
        
        if (len(all_pos[0]) <=1) & (len(all_pos[1]) >1):
            forest.nodes[all_pos[0][0]].parent = label_list[all_pos[1][0]]

        if (len(all_pos[-1]) <=1) & (len(all_pos[-2]) >1):
            forest.nodes[all_pos[-1][0]].parent = label_list[all_pos[-2][-1]]
            
        for i in range(1,len(all_pos)-1):
            if (len(all_pos[i]) <=1) & (len(all_pos[i-1])>1) & (len(all_pos[i+1])>1):
                 forest.nodes[all_pos[i][0]].parent = label_list[all_pos[i-1][-1]]

    return label_list

def preprocess2(forest,label_list,winSize): #去除突变信息，中心窗口滤波法
    tail = int((winSize-1)/2)
    for i in range(tail,len(label_list)-tail-1):
        center = label_list[i]
        neighbor = np.array(label_list)[(i-tail):(i+tail)]
        count = np.bincount(neighbor)  # 找出第3列最频繁出现的值
        maxmum = np.where(count>=tail)[0].tolist()
        
        if len(maxmum) == 0:
            continue
        if len(maxmum) == 1:
            if maxmum[0] != center:
                label_list[i] = maxmum[0]                
                forest.nodes[i].parent = label_list[i]
        if len(maxmum) == 2:
            if center in maxmum:
                1
            else:
                label_list[i] = maxmum[0]
                forest.nodes[i].parent = label_list[i]

    return label_list    
   

def get_LabelPatten(label_list): #根据label_list识别季节变化可能发生的时间段
    uniqVlaue = list(map(itemgetter(0), groupby(label_list)))
    all_pos = []
    uniqPoslist = []
    i = 0
    for j in range(len(label_list)):
        if label_list[j] == uniqVlaue[i]:
            uniqPoslist.append(j)
        else:
            all_pos.append(uniqPoslist)
            uniqPoslist = [] 
            i = i+1
            uniqPoslist.append(j)
    all_pos.append(uniqPoslist) 


    posRL_list = []
    posRL_uniqList = [] 
    firstD = 0
    pos_r_temp = []
    if len(uniqVlaue)>2:
        for i in range(2,len(uniqVlaue),1):     
            current = uniqVlaue[i]
            last1 = uniqVlaue[i-1]
            last2 = uniqVlaue[i-2]
            condition1 = ((current - last2)==0)
            if condition1:
                if firstD == 0:
                    pos_l = i-2
                    firstD = 1              
                pos_r_temp.append(i)
            else:
                if len(pos_r_temp)!=0 :
                    pos_r = pos_r_temp[len(pos_r_temp)-1]
                    firstD = 0
                    pos_r_temp = []
                    posRL_uniqList.append((pos_l,pos_r))
        if len(pos_r_temp)!=0 :
            pos_r = pos_r_temp[len(pos_r_temp)-1]
            posRL_uniqList.append((pos_l,pos_r))

        if len(posRL_uniqList) != 0:
            for pos in posRL_uniqList:
                s = all_pos[pos[0]][0]
                e = all_pos[pos[1]][-1]
                posRL_list.append((s,e))
                
    return posRL_list

def get_LabelPatten3(label_list): #根据label_list识别季节变化可能发生的时间段
    uniqVlaue = list(map(itemgetter(0), groupby(label_list)))
    all_pos = []
    uniqPoslist = []
    i = 0
    for j in range(len(label_list)):
        if label_list[j] == uniqVlaue[i]:
            uniqPoslist.append(j)
        else:
            all_pos.append(uniqPoslist)
            uniqPoslist = [] 
            i = i+1
            uniqPoslist.append(j)
    all_pos.append(uniqPoslist) 


    posRL_list = []
    posRL_uniqList = [] 
    firstD = 0
    pos_r_temp = []
    if len(uniqVlaue)>3:
        for i in range(3,len(uniqVlaue),1):     
            current = uniqVlaue[i]
            last3 = uniqVlaue[i-3]
            condition1 = ((current - last3)==0)
            if condition1:
                if firstD == 0:
                    pos_l = i-3
                    firstD = 1              
                pos_r_temp.append(i)
            else:
                if len(pos_r_temp)!=0 :
                    pos_r = pos_r_temp[len(pos_r_temp)-1]
                    firstD = 0
                    pos_r_temp = []
                    posRL_uniqList.append((pos_l,pos_r))
        if len(pos_r_temp)!=0 :
            pos_r = pos_r_temp[len(pos_r_temp)-1]
            posRL_uniqList.append((pos_l,pos_r))

        if len(posRL_uniqList) != 0:
            for pos in posRL_uniqList:
                s = all_pos[pos[0]][0]
                e = all_pos[pos[1]][-1]
                posRL_list.append((s,e))
                
    return posRL_list

def get_MSTPeriodEdge(graph_edges,period): #提取最小生成树的季节性节点数量
    G = nx.Graph()
    G.add_weighted_edges_from(graph_edges)
    T = nx.minimum_spanning_tree(G_ts)
    mst_periodEdge = []
    for (u,v,wt) in T.edges.data('weight'):
        if np.abs(int(u)-int(v)) >= period:
            if int(u)<int(v):
                mst_periodEdge.append((int(v),int(u),wt))
            else:
                mst_periodEdge.append((int(u),int(v),wt))
    return mst_periodEdge
    


#根据label获得的季节模式时间段，合并该时间段内的数据
def combination_byPeriod(forest,label_list): 
    period_pattern = get_LabelPatten(label_list)
    period_pattern3 = get_LabelPatten3(label_list)
    while len(period_pattern) != 0:
        for p in period_pattern:
            labels = label_list[p[0]:p[1]+1]
            counts = np.bincount(labels)
            setlabel = np.argmax(counts)   
            for i in range(p[0],p[1]+1):
                forest.nodes[i].parent = setlabel
        label_list = get_segLabel(forest)
        period_pattern = get_LabelPatten(label_list)
        
    if len(period_pattern3) != 0:
        for p in period_pattern3:
            labels = label_list[p[0]:p[1]+1]
            counts = np.bincount(labels)
            setlabel = np.argmax(counts)   
            for i in range(p[0],p[1]+1):
                forest.nodes[i].parent = setlabel
    return  forest





#对于非季节性数据，使用不相似矩阵，判别是否需要合并，否则不合并
def combination_byDisMatrix(forest,dissimilarMatrix,label_list,thresh_combine):
    uniqVlaue = np.unique(label_list)
    uniqVlaue = sorted(uniqVlaue)  #节点排列
    comb = list(combinations(uniqVlaue, 2)) #节点组合  
    if len(uniqVlaue) != 1:
        for i in range(len(comb)):
            if comb[i][1] != comb[i][0]:
                pl_node = np.array(forest.findNodesbyParent(comb[i][0]))
                pr_node = np.array(forest.findNodesbyParent(comb[i][1]))

                dis = dissimilarMatrix[pr_node,:][:,pl_node]

                value = np.mean(dis) #暂时使用均值
                condition = (value<=thresh_combine)
                if condition:
                    if np.size(pl_node)<=np.size(pr_node):
                        for n in pl_node.tolist():
                            forest.nodes[n].parent = comb[i][1]
                        # comb更新
                        temp = np.array(comb).astype(np.int8)
                        temp[np.where(temp==comb[i][0])] = comb[i][1]
                        comb = list(map(tuple,temp))
                    else:
                        for n in pr_node.tolist():
                            forest.nodes[n].parent = comb[i][0]
                        # comb更新
                        temp = np.array(comb).astype(np.int8)
                        temp[np.where(temp==comb[i][1])] = comb[i][0]
                        comb = list(map(tuple,temp))               
    return forest










def build_graph_ts(similar_matrix, ts_length, neighbor,lag,interval,period, period_neighbor):
    graph_edges = []
    G_ts = nx.Graph()
   
    for t in range(ts_length):

        # 邻域模式
        if neighbor !=0 :
            for n in range(0,neighbor):
                if t-n > 0:
                    edge = (t,t-n-1,1-similar_matrix[t,t-n-1])
                    graph_edges.append(edge)
                    G_ts.add_edge(str(t), str(t-n-1), weight=1-similar_matrix[t,t-n-1])
        # lag 模式
        if lag != 0:
            if t-lag >= 0:
                edge = (t,t-lag,1-similar_matrix[t,t-lag])
                graph_edges.append(edge)
                G_ts.add_edge(str(t), str(t-lag), weight = 1-similar_matrix[t,t-lag])
        
        #周期模式
        if period != 0:
            if (t + period) <  ts_length:
                edge = (t + period,t,1-similar_matrix[t + period,t])
                graph_edges.append(edge)
                G_ts.add_edge(str(t + period), str(t), weight=1-similar_matrix[t + period,t])
                if period_neighbor !=0:
                    for n in range(period_neighbor):
                        if t + period-n > 0:
                            edge = (t ,t + period-n-1,1-similar_matrix[t ,t + period-n-1])
                            graph_edges.append(edge)
                            G_ts.add_edge(str(t), str(t + period-n-1), weight=1-similar_matrix[t ,t + period-n-1])
        
        
    # interval 模式
    if len(interval) != 0:
        for ins in interval: 
            u = 0
            for t in range(ts_length):  
                if ins != 0:
                    if (t-ins) == u*ins:
                        edge = (t, t-ins,1-similar_matrix[t, t-ins])
                        graph_edges.append(edge)
                        G_ts.add_edge(str(t), str( t-ins), weight = 1-similar_matrix[t, t-ins])
                        u = u+1
                            
    return graph_edges,G_ts

def get_breaks(forest):
    #breaks = []
    breakpos = []
    
    label_list = get_segLabel(forest)
    uniqVlaue = list(map(itemgetter(0), groupby(label_list)))
    u = np.unique(np.array(label_list)).tolist()
    
     
    all_pos = []
    uniqPoslist = []
    i = 0
    for j in range(len(label_list)):
        if label_list[j] == uniqVlaue[i]:
            uniqPoslist.append(j)
        else:
            all_pos.append(uniqPoslist)
            uniqPoslist = [] 
            i = i+1
            uniqPoslist.append(j)
    all_pos.append(uniqPoslist) 
    
    
    if len(all_pos) > 1:
        for i in range(1,len(all_pos)):
            #breaks.append(ts_date[all_pos[i][0]])
            breakpos.append(all_pos[i][0])
            
    
    if len(u) < len(uniqVlaue):
        return breakpos,['可能存在季节周期'],pd.DataFrame({'forest':forest.nodes})
    
    else:
        return breakpos,['正常'],pd.DataFrame({'forest':forest.nodes})


    
################ 必要函数定义
global diff
def diff(img, x1, y1, x2, y2):
    _out = np.sum((img[x1, y1] - img[x2, y2]) ** 2)
    return np.sqrt(_out)

global threshold_func
def threshold_func(size, const):
    return (const * 1.0 / size)

    
def graphSegMerge(similar_matrix,NL,PL,K,M):
    
    ################ 参数设置
    similar_matrix = 1-similar_matrix
    ts_length = similar_matrix.shape[0]
    
    # 论文中提到的参数
    neighbor = NL #
    period = PL
    percent = K
    thresh_combine = M
    
   
    # 辅助参数：
    period_neighbor = 1 #一般为0-2，与图构建相关
    K = 0.64 #不设置也可以，与percent共同控制分割情况
    min_comp_size = 3   # 一般为2-3，与thresh_combine共同控制合并情况
     
    
    # 没用的参数：
    percent_thre = 0
    lag = 0
    interval = []
    
    
    
    
    
    ################### 图构建
    graph_edges,G_ts = build_graph_ts(similar_matrix, ts_length, neighbor,lag ,interval, period, period_neighbor)

    
    
    
    ################# 阈值测定
    weight = lambda edge: edge[2]
    sorted_graph = sorted(graph_edges, key=weight)
    #sorted_graph_list.append(sorted_graph)
    
    K1 = np.percentile(np.array(sorted_graph)[:,2],percent)
    if K1<K:
        K1 = K
    #print(i,K1)
    
    #newgraph
    ss = np.array(sorted_graph)
    dis = ss[:,0]-ss[:,1]
    temp = np.vstack((ss[np.where(np.abs(dis)<=10)[0],:],ss[np.where(np.abs(dis)>10)[0],:]))
    newgraph = []
    for u in temp:
        newgraph.append((int(u[0]),int(u[1]),u[2]))
    
    thresh_combine1 = np.percentile(np.array(sorted_graph)[:,2].flatten(),percent_thre)
    if thresh_combine1 < thresh_combine:
        thresh_combine1 = thresh_combine
    #thresh_combine1 = np.sum(np.array(sorted_graph)[:,2].flatten())/len(np.array(sorted_graph)[:,2].flatten())
        
    
    
    
    #############图分割和融合
    graph_edges = newgraph
    num_nodes = ts_length
    const = K1
    min_size = min_comp_size
   
    similar_matrix = similar_matrix
    thresh_combine = thresh_combine1
    
    
    # Step 1: initialization
    forest = Forest(num_nodes)
    weight = lambda edge: edge[2]
    #sorted_graph = sorted(graph_edges, key=weight)
    threshold = []
    for i in range(num_nodes):
        threshold.append((const * 1.0 / 1))

    # Step 2: merging
    for edge in sorted_graph:
        parent_a = forest.find(edge[0])
        parent_b = forest.find(edge[1])
        a_condition = weight(edge) <= threshold[parent_a]
        b_condition = weight(edge) <= threshold[parent_b]

        if parent_a != parent_b and a_condition and b_condition:
            forest.merge(parent_a, parent_b)
            a = forest.find(parent_a)
            threshold[a] = weight(edge) + (const * 1.0 / forest.nodes[a].size)

    # Step 3: 原文中的合并方法
    remove_small_components(forest, sorted_graph, min_size)

    # Step 4：我们的合并
    


    # 季节合并：
    label_list = get_segLabel(forest)
    combination_byPeriod(forest,label_list)
    
    # 噪声去除
    label_list = get_segLabel(forest)
    preprocess(forest,label_list)
    
    # 阈值合并：
    label_list = get_segLabel(forest)
    combination_byDisMatrix(forest,1-similar_matrix,label_list,thresh_combine)
     
        
        
    ################## 结果提取
    pos,breaks,rlist = get_breaks(forest)    
        
    
    return pos


```

# 函数3：精度评价

```python

def spatial_ac(ref_breaksList,est_breaksList):
    
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import confusion_matrix

    
    ref_ifchange = []
    est_ifchange = []
    for i in range(len(ref_breaksList)):
        if len(ref_breaksList[i])==0:
            ref_ifchange.append(0)
        else:
            ref_ifchange.append(1)
        if len(est_breaksList[i])==0:
            est_ifchange.append(0)
        else:
            est_ifchange.append(1)

    confusion_matrix = confusion_matrix(np.array(ref_ifchange), np.array(est_ifchange))
    #print('SPA_confusion_matrix:')
    #print('unchanged  change')
    #print(confusion_matrix)


    SPA_UA = confusion_matrix[1,1]/(confusion_matrix[0,1]+confusion_matrix[1,1])
    #print('SPA_UA: ',SPA_UA)

    SPA_PA = confusion_matrix[1,1]/(confusion_matrix[1,0]+confusion_matrix[1,1])
    #print('SPA_PA: ',SPA_PA)
    

    TP = confusion_matrix[1,1]
    FP = confusion_matrix[1,0]
    FN = confusion_matrix[0,1]
    TN = confusion_matrix[0,0]
    P = TP/(TP+FP)
    R = TP/(TP+FN)
    SPA_F1 = 2*P*R/(P+R)
    #print('SPA_F1: ',SPA_F1)
    
    confm = np.array([[confusion_matrix[1,1],confusion_matrix[1,0]],[confusion_matrix[0,1],confusion_matrix[0,0]]])
    
    return confm, SPA_UA, SPA_PA, SPA_F1


def temporal_ac(ref_breaksList,est_breaksList):
    furry = 2

    ref_ifchange = []
    est_ifchange = []
    for i in range(len(ref_breaksList)):
        if len(ref_breaksList[i])==0:
            ref_ifchange.append(0)
        else:
            ref_ifchange.append(1)
        if len(est_breaksList[i])==0:
            est_ifchange.append(0)
        else:
            est_ifchange.append(1)

    ref = []
    bre = []        
    for i in range(len(ref_breaksList)):
        if (ref_ifchange[i]== 1) & (est_ifchange[i]==1):
            ref.append(ref_breaksList[i])
            bre.append(est_breaksList[i])


    from bisect import bisect_left
    def takeClosest(myList, myNumber):
        if myNumber >= max(myList):
            return max(myList)
        elif myNumber <= min(myList):
            return min(myList)
        if myList[0] < myList[1]:
            pos = bisect_left(myList, myNumber)   # 找到 mylist 里面第一个不比 mynumber 小（即 >= ）的数的索引下标
            # 返回的插入点 pos 可以将数组myList分成两部分。左侧是 all(val < x for val in myList[lo:i]) ，右侧是 all(val >= x for val in myList[i:hi])
            before = myList[pos - 1]
            after = myList[pos]
            if after - myNumber < myNumber - before:
                return after
            else:
                return before
        else:
            myList = myList[::-1]
            pos = takeClosest(myList, myNumber)
            return pos

    CC = 0 # 参考：变化， 探测：变化
    CN = 0 # 参考：变化， 探测：无变化
    NC = 0 # 参考：无变化  探测：变化
    NN = 0 # 参考：无变化  探测：无变化
    for i in range(len(ref)):
        if (len(ref[i]) != 0) & (len(bre[i]) != 0):
            DCC = 0
            curr = 0
            for j in range(len(ref[i])):
                t = takeClosest(bre[i], ref[i][j])
                if np.abs(t- ref[i][j]) <= furry:
                    DCC = DCC + 1   
                    curr = curr + 1
                    if curr>= len(bre[i]):
                        break
            CN = CN + (len(ref[i])- DCC)
            NC = NC + (len(bre[i])- DCC)
            NN = NN + len(dateMonth) - (len(ref[i])- DCC) - (len(bre[i])- DCC) - DCC
            CC = CC + DCC

        if (len(ref[i]) != 0) & (len(bre[i]) == 0):
            CN = CN + (len(ref[i]))
            NN = NN + len(dateMonth) - (len(ref[i]))        

        if (len(ref[i]) == 0) & (len(bre[i]) != 0):
            NC = NC + (len(bre[i]))
            NN = NN + len(dateMonth) - (len(bre[i]))         

        if (len(ref[i]) == 0) & (len(bre[i]) == 0):      
            NN = NN + len(dateMonth)     

    confm = np.array([[CC,CN],[NC,NN]])
    #print('confm:      ')
    #print(confm)


    TEM_UA = confm[0,0]/(confm[0,0]+confm[1,0])
    #print('TEM_UA:   ',TEM_UA)

    TEM_PA = confm[0,0]/(confm[0,0]+confm[0,1])
    #print('TEM_PA:   ',TEM_PA)


    TP = confm[0,0]
    FP = confm[0,1]
    FN = confm[1,0]
    TN = confm[1,1]
    P = TP/(TP+FP)
    R = TP/(TP+FN)
    TEM_F1 = 2*P*R/(P+R)
    #print('TEM_F1:    ',TEM_F1)
    return confm, TEM_UA, TEM_PA, TEM_F1



def accuracyAssessment(ref_breaksList,est_breaksList):
    SPA_confm, SPA_UA, SPA_PA, SPA_F1 = spatial_ac(ref_breaksList,est_breaksList)
    TEM_confm, TEM_UA, TEM_PA, TEM_F1 = temporal_ac(ref_breaksList,est_breaksList)
    print('SPA_confusion_matrix:')
    print(SPA_confm)
    print('SPA_UA： ',SPA_UA)
    print('SPA_PA： ',SPA_PA)
    print('SPA_F1： ',SPA_F1)
    print(  '         ' )
    print('TEM_confusion_matrix:')
    print(TEM_confm)
    print('TEM_UA： ',TEM_UA)
    print('TEM_PA： ',TEM_PA)
    print('TEM_F1： ',TEM_F1)
```

# 类管理器-本地用

```python
class dataManager:
    
    def __init__(self):   
        
        #ee_feature = ee.Feature(ee_feature)     
        # 初始化属性信息
        self.propertyNames = []#,'ee_patch']#,'patchLon','patchLat']
        self.propertyVlalues = []#[name,ee_patch]#,lon,lat]
     
    ################################## 属性操作
    # 设置属性
    def set(self, str_property, list_propertyValue):
        if str_property not in self.propertyNames:
            self.propertyNames.append(str_property)
            self.propertyVlalues.append(list_propertyValue)
        if str_property in self.propertyNames:
            index = self.propertyNames.index(str_property)
            self.propertyVlalues[index] = list_propertyValue
    
    
    # 获得属性名称列表
    def getPropertyNames(self):
        return self.propertyNames
    
    
    # 获得属性值
    def get(self,str_property):
        if str_property in self.propertyNames:
            index = self.propertyNames.index(str_property)
            result  = self.propertyVlalues[index]
        if str_property not in self.propertyNames:
            result  = []
        return result
    
    
# 管理 ChangeDetection 对象构建的列表
class ObjectSummary:
    
    def __init__(self,objList):     
        self.columns = objList[0].getPropertyNames()
        self.objList = objList
       
    # 获得对象列表的全部某一属性值，返回列表
    def get(self,str_property):            
        results = []
        for obj in self.objList:           
            if str_property in self.columns:
                result  = obj.get(str_property)
            if str_property not in self.columns:
                result  = 'noll'
            results.append(result)           
        return results
    
    
    # 获得对象列表的全部某些属性值（原属性值为列表类型），返回字典
    def get2(self, list_property):
        dic = {}
        for obj in self.objList:
            key = obj.get('id')           
            for p in list_property:
                r = obj.get(p)
                keyt = str(key) + p 
                dic[keyt] = r
        return dic
    
    # 获得对象列表的全部某些属性值（原属性值为数值），返回字典
    def get1(self, list_property):
        dic = {}
        for p in list_property:
            r = self.get(p)
            dic[p] = r  
        return dic

# 多个 “时间段不同、patch数量相同”的的objlist放入一个列表后合并，返回一个objlist
def objectsCombine(multiObj_List,bandList):
    if len(multiObj_List)<= 1:
        return 0
    
    # 判断待合并的objList是不是等长（即每个objlist包含的patchList都相同）
    length = []
    for objList in multiObj_List:
        length.append(len(objList))
    lenUniq = np.unique(np.array(length))
    if np.size(lenUniq) > 1:
        return 0 
    
    # 数据合并
    length = len(multiObj_List[0])
    re = multiObj_List[0]
    for  i in range(length):
        for bandname in bandList:
            value = re[i].get(bandname+'_value')
            for objList in multiObj_List[1:]:
                value = np.dstack ((value,objList[i].get(bandname+'_value')))
            re[i].set(bandname+'_value',value)
    
    # 时间合并   
    for i in range(length):
        datelist = multiObj_List[0][i].get('dateList')
        for objList in multiObj_List[1:]:
            datelist = datelist+ objList[i].get('dateList')
        re[i].set('dateList',datelist)
    return re



# 提取objlist中指定时间段的波段数据，返回等长度objlist
def objectExtractbyDate(objList,bandList,date_pos):
    re = objList
    for bandname in bandList:
        for obj in objList:
            re[i].set(bandname+'_value',obj.get(bandname+'_value')[:,:,np.array(date_pos)])
            
    for i in range(len(objList)):
        datelist = np.array(objList[i].get('dateList'))[np.array(date_pos)].tolist()
        re[i].set('dateList',datelist)
    return re
        
    
```

```python
################################## 根据 patch 从GEE提取numpy数据
#根据波段名称，将每个patch内的数据转为float64—numpy，空值为NAN
def getPatch_imageValue(ee_imageCollection,ee_feature,str_bandName):
    ee_feature = ee.Feature(ee_feature)
    ee_geometry = ee_feature.geometry()
    def extractSameband(list1):
        mm = maskInside(ee.Image(list1).select(str_bandName),ee_geometry)
        test = ee.Image(mm).sampleRectangle(**{'region':ee_geometry})
        return test.getArray(str_bandName)
    uu = ee_imageCollection.toList(500).map(extractSameband)
    nochange_pixels = np.array(ee.Array.cat(uu,2).getInfo())
    nochange_pixels = nochange_pixels.astype('float64')
    nanpos_notC = (nochange_pixels==0)
    nochange_pixels[nanpos_notC] = np.NaN
    return nochange_pixels



#掩膜边界之外的影像数据
def maskInside(ee_image,ee_geometry):
    mask = ee.Image.constant(1).clip(ee_geometry).mask()
    masknot = ee.Image.constant(1).clip(ee_geometry).mask().Not()
    back = ee_image.updateMask(masknot).neq(ee_image.updateMask(masknot))
    return ee.ImageCollection([ee_image.updateMask(mask),back]).mosaic().reproject(ee_image.select(0).projection())

 
```
